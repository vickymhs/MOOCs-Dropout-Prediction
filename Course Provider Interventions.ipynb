```python
dfpred=dfpred.drop(columns='Unnamed: 0')
```

```python
dfpred=dfpred.values
```

```python
predval=[]
predval.append(dfpred[0][0])
predval.append(dfpred[1][0])
predval.append(dfpred[2][0])
predval.append(dfpred[3][0])
predval.append(dfpred[4][0])
```

```python
predval[0]
```

```python
fig, ax = plt.subplots()
low = min(predval)
high = max(predval)
plt.ylim([math.ceil(low-0.5*(high-low)), math.ceil(high+0.5*(high-low))])

label=['Week 1','Week 2', 'Week 3', 'Week 4', 'Week 5']
index = np.arange(len(label))
plt.bar(index, predval)
plt.xlabel('Weeks', fontsize=10)
plt.ylabel('No of Dropouts', fontsize=10)
plt.xticks(index, label, fontsize=5, rotation=30)
plt.title('Dropouts in entire course')
plt.show()
```
